```python
import pandas as pd
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
from future.utils import iteritems
from collections import Counter
from sklearn.manifold import TSNE
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
import pymysql
```

```python
article_data = pd.read_csv('mixver8.txt', encoding='utf-8', header= None)
documents = [' '.join(i[0].split(' ')[1:]) for i in article_data.values]
```

```python
print(documents[0])
```

```python
as_one = ''
for document in documents:
    as_one = as_one + ' ' + document
words = as_one.split()
```

```python
words[0:10]
```

```python
counts = Counter(words)
counts
```

```python
vocab = sorted(counts, key=counts.get, reverse=True)
vocab
```

```python
word2idx = {word.encode("utf8").decode("utf8"): ii for ii, word in enumerate(vocab,1)}
```

```python
word2idx
```

```python
idx2word = {ii: word for ii, word in enumerate(vocab)}
idx2word
```
